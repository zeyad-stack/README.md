```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# 1Ô∏è‚É£ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿµŸÅÿ≠ÿ©
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}  # ÿ®ÿπÿ∂ ÿßŸÑŸÖŸàÿßŸÇÿπ ÿ™ÿ≠ÿ¨ÿ® ÿßŸÑŸàŸäÿ® ÿ≥ŸÉÿ±ÿßÿ®ŸäŸÜÿ∫ÿå ŸÑÿ∞ÿß ŸÜÿ≥ÿ™ÿÆÿØŸÖ ŸáŸäÿØÿ± ŸàŸáŸÖŸä
response = requests.get(url, headers=headers)

# 2Ô∏è‚É£ ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿµŸÅÿ≠ÿ© ÿ•ŸÑŸâ ŸÉÿßÿ¶ŸÜ BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# 3Ô∏è‚É£ ÿßŸÑÿ®ÿ≠ÿ´ ÿπŸÜ ÿ¨ŸÖŸäÿπ ÿßŸÑÿ¨ÿØÿßŸàŸÑ ŸÅŸä ÿßŸÑÿµŸÅÿ≠ÿ©
tables = soup.find_all("table")

# 4Ô∏è‚É£ ÿ∑ÿ®ÿßÿπÿ© ÿ¨ŸÖŸäÿπ ÿßŸÑÿ¨ÿØÿßŸàŸÑ ŸÑÿßŸÉÿ™ÿ¥ÿßŸÅ ŸÖŸÉÿßŸÜ ÿßŸÑÿ¨ÿØŸàŸÑ ÿßŸÑÿµÿ≠Ÿäÿ≠
for index, table in enumerate(tables):
    print(f"üìå Table {index}:")
    print(table.text[:500])  # ÿ∑ÿ®ÿßÿπÿ© ÿ¨ÿ≤ÿ° ŸÖŸÜ ŸÖÿ≠ÿ™ŸàŸâ ÿßŸÑÿ¨ÿØŸàŸÑ ŸÑŸÅŸáŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
    print("=" * 50)

# 5Ô∏è‚É£ ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ¨ÿØŸàŸÑ ÿßŸÑÿµÿ≠Ÿäÿ≠ ŸäÿØŸàŸäŸãÿß ÿ•ÿ∞ÿß ÿ™ŸÖ ÿßŸÑÿπÿ´Ÿàÿ± ÿπŸÑŸäŸá
# ÿßÿ≥ÿ™ÿ®ÿØŸÑ `tables[X]` ÿ®ÿßŸÑÿ¨ÿØŸàŸÑ ÿßŸÑÿµÿ≠Ÿäÿ≠ ÿ®ÿπÿØ ÿßŸÑŸÖÿ±ÿßÿ¨ÿπÿ©
revenue_table = tables[0]  # ÿ¨ÿ±Ÿëÿ® ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑÿ±ŸÇŸÖ ÿ®ÿπÿØ ŸÖÿ±ÿßÿ¨ÿπÿ© ÿßŸÑÿ∑ÿ®ÿßÿπÿ©

# 6Ô∏è‚É£ ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
rows = revenue_table.find_all("tr")
data = []
for row in rows[1:]:  # ÿ™ÿÆÿ∑Ÿä ÿßŸÑÿπŸÜŸàÿßŸÜ
    cols = row.find_all("td")
    if len(cols) == 2:
        year = cols[0].text.strip()
        revenue = cols[1].text.strip()
        data.append([year, revenue])

# 7Ô∏è‚É£ ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ DataFrame
tesla_revenue = pd.DataFrame(data, columns=["Year", "Revenue"])

# 8Ô∏è‚É£ ÿπÿ±ÿ∂ ÿ¢ÿÆÿ± 5 ÿµŸÅŸàŸÅ
print(tesla_revenue.tail())

# 9Ô∏è‚É£ ÿ≠ŸÅÿ∏ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÅŸä ŸÖŸÑŸÅ CSV
tesla_revenue.to_csv("tesla_revenue.csv", index=False)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# 1Ô∏è‚É£ ÿ™ÿ≠ŸÖŸäŸÑ ÿµŸÅÿ≠ÿ© ÿßŸÑÿ•Ÿäÿ±ÿßÿØÿßÿ™ ŸÖŸÜ ŸÖŸàŸÇÿπ Macrotrends
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
headers = {"User-Agent": "Mozilla/5.0"}  # ÿ™ÿ¨ŸÜÿ® ÿßŸÑÿ≠ÿ∏ÿ± ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ User-Agent
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# 2Ô∏è‚É£ ÿßŸÑÿ®ÿ≠ÿ´ ÿπŸÜ ÿßŸÑÿ¨ÿØŸàŸÑ ÿßŸÑÿµÿ≠Ÿäÿ≠ ŸÅŸä ÿßŸÑÿµŸÅÿ≠ÿ©
tables = soup.find_all("table")
revenue_table = None

for table in tables:
    if "GameStop Annual Revenue" in table.text:  # ÿßŸÑÿ®ÿ≠ÿ´ ÿπŸÜ ÿßŸÑÿ¨ÿØŸàŸÑ ÿßŸÑŸÖÿ∑ŸÑŸàÿ®
        revenue_table = table
        break

# 3Ô∏è‚É£ ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÖŸÜ ÿßŸÑÿ¨ÿØŸàŸÑ
data = []
if revenue_table:
    rows = revenue_table.find_all("tr")
    for row in rows[1:]:  # ÿ™ÿÆÿ∑Ÿä ÿßŸÑÿπŸÜŸàÿßŸÜ
        cols = row.find_all("td")
        if len(cols) == 2:
            year = cols[0].text.strip()
            revenue = cols[1].text.strip()
            data.append([year, revenue])

# 4Ô∏è‚É£ ÿ•ŸÜÿ¥ÿßÿ° DataFrame
gme_revenue = pd.DataFrame(data, columns=["Year", "Revenue"])

# 5Ô∏è‚É£ ÿπÿ±ÿ∂ ÿ¢ÿÆÿ± 5 ÿµŸÅŸàŸÅ
print(gme_revenue.tail())

# 6Ô∏è‚É£ ÿ≠ŸÅÿ∏ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÅŸä ŸÖŸÑŸÅ CSV
gme_revenue.to_csv("gme_revenue.csv", index=False)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿ≥ŸáŸÖ ÿ™ÿ≥ŸÑÿß
tesla_data = yf.download("TSLA", start="2010-01-01", end="2025-01-01")

# 2Ô∏è‚É£ ÿØÿßŸÑÿ© ŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
def make_graph(stock_data, title="Stock Price Graph"):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data["Close"], label="Closing Price", color="blue")
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# 3Ô∏è‚É£ ÿ±ÿ≥ŸÖ ŸÖÿÆÿ∑ÿ∑ ÿ≥ŸáŸÖ ÿ™ÿ≥ŸÑÿß ÿ®ÿπŸÜŸàÿßŸÜ ŸÖÿÆÿµÿµ
make_graph(tesla_data, title="Tesla Stock Price Over Time")
```
